```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("Facebook Friends.xlsx")
```

```python
for col in df.columns:
    df[col] = df[col].astype(str).str.replace(",", "")
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
print("Column Names:")
print(df.columns)
```

```python
print("\nFirst 5 Rows:")
print(df.head())
```

```python
print("\nLast 5 Rows:")
print(df.tail())
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
plt.figure()
df['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
df['Friends'].hist(bins=30)
plt.title("Distribution of Facebook Friends")
plt.xlabel("Number of Friends")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
df['Posts'].hist(bins=30)
plt.title("Distribution of Posts")
plt.xlabel("Number of Posts")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
df['Gender '].value_counts().plot(kind='bar')
plt.title("Gender Frequency")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
corr = df.corr()
plt.imshow(corr)
plt.colorbar()
plt.title("Correlation Matrix")
plt.show()
```
